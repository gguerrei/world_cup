```python
import pdfplumber
import pandas as pd
```

```python
pdf = pdfplumber.open('/Users/Gui/Desktop/code/world_cup_code/PDF_stadium_usage/ARENA PANTANAL/combined_pantanal.pdf')
```

```python
first_game = pdf.pages[0]
first_game_text = first_game.extract_text()
print(first_game_text)
```

```python
first_game_text_clean = first_game_text.split('\n')
first_game_text_clean
```

```python
free_tickets = []
for sentences in first_game_text_clean:
    if 'CORTESIA' in sentences:
        free_tickets.append(sentences)
    if 'GRATUIDADE' in sentences:
        free_tickets.append(sentences)

free_tickets
```

```python
free_tickets_numbers = [sentence.split()[-3] for sentence in free_tickets]
free_tickets_numbers
```

```python
for numbers in range(len(free_tickets_numbers)):
    free_tickets_numbers[numbers] = int(free_tickets_numbers[numbers])
```

```python
free_tickets_numbers_list = [sum(free_tickets_numbers)]
free_tickets_numbers_list
```

```python
for i in range(2, 38, 2):
    game = pdf.pages[i]
    game_text = game.extract_text()
    game_text_clean = game_text.split('\n')
    free_tickets_buffer = []
    for sentences in game_text_clean:
        if 'CORTESIA' in sentences:
            free_tickets_buffer.append(sentences)
        if 'GRATUIDADE' in sentences:
            free_tickets_buffer.append(sentences)
    free_tickets_buffer = [sentence.split()[-3] for sentence in free_tickets_buffer]
    for numbers in range(len(free_tickets_buffer)):
        free_tickets_buffer[numbers] = free_tickets_buffer[numbers].replace('.', '')
        free_tickets_buffer[numbers] = int(free_tickets_buffer[numbers])
    free_tickets_buffer = sum(free_tickets_buffer)
    free_tickets_numbers_list.append(free_tickets_buffer)
free_tickets_numbers_list
```

```python
games_id = ["Nº 14155", "Nº 14194", "Nº 14301", "Nº 14454", "Nº 14658", "Nº 14692", "Nº 14750", "Nº 14840", "Nº 14914", "Nº 14970", "Nº 15052", "Nº 15086", "Nº 15104", "Nº 15127", "Nº 15128", "Nº 15149", "Nº 15161", "Nº 15182", "Nº 15191"]
```

```python
games_id_free_tickets = {games_id[i]: free_tickets_numbers_list[i] for i in range(len(games_id))}
games_id_free_tickets
```

```python
df = pd.DataFrame(list(games_id_free_tickets.items()), columns = ['Game ID', 'Free Tickets'])
df
```

```python
first_game_tables = first_game.extract_table()
total_tickets = first_game_tables[24]
total_tickets[4] = total_tickets[4].replace('.', '')
total_tickets_list = [int(total_tickets[4])]
total_tickets_list
```

```python
for i in range(2, 38, 2):
    game = pdf.pages[i]
    game_tables = game.extract_table()
    game_tables_correct = len(game_tables) - 1
    total_tickets = game_tables[game_tables_correct]
    total_tickets[4] = total_tickets[4].replace('.', '')
    total_tickets_list.append(int(total_tickets[4]))
total_tickets_list
```

```python
df['Total Tickets'] = total_tickets_list
df 
```

```python
df['Percentage of Free Tickets'] = (df['Free Tickets'] / df['Total Tickets'])*100
df
```
